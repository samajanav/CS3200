## Visualization part of the assignment

```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('results.csv')
df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(10, 6))
plt.barh(df['gene'], df['num_records'], color='teal')
plt.xlabel('Number of Positive Associations')
plt.title('Top 5 Genes with the Most Positive Associations')
plt.gca().invert_yaxis()
plt.tight_layout()

# Show the chart
plt.show()
```
